```python
import pandas as pd
```

```python
residencial = pd.read_csv('alquiler_residencial.csv')
residencial.head(30)
```

```python
list(residencial['Tipo'].drop_duplicates())
```

```python
sel_depart = residencial['Tipo'] == 'Departamento'
cant_dept = residencial[sel_depart].shape[0]
cant_dept
```

```python
sel_chc = (residencial['Tipo']== 'Casa')|(residencial['Tipo']== 'Habitación')|(residencial['Tipo']== 'Cochera')
cant_chc = residencial[sel_chc].shape[0]
cant_chc
```

```python
sel_habxprecio = (residencial['Cuartos']>= 4) & (residencial['Valor']<=  2000)
cant_habxprecio = residencial[sel_habxprecio].shape[0]
cant_habxprecio
```

```python
sel_areas = (residencial['Area']>= 60) & (residencial['Area']<=  100)
cant_areas = residencial[sel_areas].shape[0]
cant_areas
```

```python
print(f'La cantidad de departamentos encontrados es de {cant_dept}')
print(f'La cantidad de casas, habitaciones y cocheras encontradps es de {cant_chc}')
print(f'La cantidad de casas de 4 o más habitaciones que cuestan menos de 2000 es {cant_habxprecio}')
print(f'La cantidad de vivendas con un area entre 60 y 100 mts cuadrados es de {cant_areas}')
```
